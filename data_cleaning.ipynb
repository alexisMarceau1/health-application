```python
# fonction pour l'équivalence d'un dictionnaire
def equivalence(produit):
    if produit in en_to_fr.keys():
        return en_to_fr[produit]
    return produit
```

Je drop aussi `fiber_100g` qui ne sera pas utile

```python
df_fr = df_fr.drop(columns=['energy_100g'])
df_fr = df_fr.drop(columns=['fiber_100g'])
```

Je renomme la varibale du nutriscore pour la séparer des autres variables

```python
df_fr.rename(columns = {"nutrition-score-fr_100g":"nutrition_score"}, inplace=True)
```

Je sélectionne les variables finissant par _100g pour les traiter ensembles

```python
per_100 = df_fr.filter(regex='_100g$',axis=1)
```

**La valeur d'un macronutriment doit être compris entre 0 et 100**

```python
df_fr.drop(per_100[( (per_100 > 100) | (per_100 < 0)).any(axis=1)].index, inplace=True)
```

```python
print("Dimensions : ", df_fr.shape)
```

**La somme des macronutriments doit être < 100**

```python
df_fr.drop(df_fr[ (df_fr['fat_100g'] + df_fr['carbohydrates_100g'] + df_fr['proteins_100g'] > 100) ].index, inplace=True)
```

```python
print("Dimensions : ", df_fr.shape)
```

**La valeur des nutriments enfant doit être inférieure à celle des nutriments parent**

```python
df_fr.drop(df_fr[(df_fr['saturated-fat_100g']> df_fr['fat_100g'])].index, inplace=True)
df_fr.drop(df_fr[(df_fr['sugars_100g']> df_fr['carbohydrates_100g'])].index, inplace=True)
```

```python
print("Dimensions : ", df_fr.shape)
```

**J'utilise l'écart interquartile par catégories pour supprimer de potentiels outliers**

```python
nutriments = ["fat_100g", "carbohydrates_100g", "proteins_100g", "salt_100g", "sodium_100g","sugars_100g","saturated-fat_100g"]
```

```python
def quantile_condition(group):
    Q1 = group.quantile(0.25)
    Q3 = group.quantile(0.75)
    IQR = Q3 - Q1
    condition = ~((group < (Q1 - 1.5 * IQR)) | (group > (Q3 + 1.5 * IQR)))
    return condition
```

```python
grouped = df_fr.groupby(["main_category_fr"])[nutriments]
conditions = grouped.transform(quantile_condition)
df_fr = df_fr[conditions.all(axis=1)]
```
